```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# 1. LOAD DATA

df = pd.read_csv("AirQuality.csv", sep=';', decimal=',')

# Xóa cột rỗng
df = df.drop(columns=['Unnamed: 15', 'Unnamed: 16'], errors='ignore')

print("=== ORIGINAL SHAPE ===")
print(df.shape)

# 2. EDA + CLEANING

# Thay -200 bằng NaN
df.replace(-200, np.nan, inplace=True)

print("\n=== MISSING VALUES BEFORE ===")
print(df.isnull().sum())

# Xử lý missing values
for col in df.columns:
    if df[col].dtype != 'object':
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])

print("\n=== MISSING VALUES AFTER ===")
print(df.isnull().sum().sum())

# 3. OUTLIER HANDLING (IQR)

numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

print("\n=== OUTLIERS AFTER ===")
print(df.isnull().sum().sum())
```

```python
# 1. ENCODING / FEATURE ENGINEERING

# Chuyển Date + Time sang datetime
df['Datetime'] = pd.to_datetime(
    df['Date'] + ' ' + df['Time'],
    format='%d/%m/%Y %H.%M.%S'
)

# Tách thành feature thời gian
df['Day'] = df['Datetime'].dt.day
df['Month'] = df['Datetime'].dt.month
df['Year'] = df['Datetime'].dt.year
df['Hour'] = df['Datetime'].dt.hour

# Xóa cột không cần
df = df.drop(columns=['Date', 'Time', 'Datetime'])

# 2. FEATURE SELECTION

df = df.drop(columns=['NMHC(GT)'], errors='ignore')
```

```python
# 1. SCALING

scaler = StandardScaler()

numeric_cols = df.columns
df_scaled = pd.DataFrame(
    scaler.fit_transform(df),
    columns=numeric_cols
)

# 7. OUTPUT

print("\n=== DATA AFTER SCALING ===")
print(df_scaled.head())

print("\n=== FINAL SHAPE ===")
print(df_scaled.shape)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# ======================
# 0. CHUẨN BỊ DỮ LIỆU
# ======================
target_col = 'CO(GT)'

y = df[target_col]
X = df.drop(columns=[target_col])

print("X shape:", X.shape)
print("y shape:", y.shape)

# ======================
# 1. CHUẨN HÓA DỮ LIỆU
# ======================
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ======================
# 2. PCA GIẢM CHIỀU
# ======================
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

print("\nExplained Variance Ratio:")
print(pca.explained_variance_ratio_)

# ======================
# 3. CHỌN K BẰNG BIC
# ======================
bic_scores = []
sil_scores = []

k_range = range(2, 7)

for k in k_range:
    gmm = GaussianMixture(
        n_components=k,
        covariance_type='full',
        random_state=42
    )

    gmm.fit(X_pca)

    # BIC
    bic = gmm.bic(X_pca)
    bic_scores.append(bic)

    # Silhouette
    labels = gmm.predict(X_pca)
    sil = silhouette_score(X_pca, labels)
    sil_scores.append(sil)

# K tối ưu theo BIC (BIC thấp nhất)
optimal_k = k_range[np.argmin(bic_scores)]

print("\n=== BIC RESULTS ===")
for i, k in enumerate(k_range):
    print(
        f"k={k} | "
        f"BIC={bic_scores[i]:.2f} | "
        f"Silhouette={sil_scores[i]:.4f}"
    )

print("\nOptimal K (BIC):", optimal_k)

# ======================
# 4. HUẤN LUYỆN GMM CUỐI
# ======================
best_gmm = GaussianMixture(
    n_components=optimal_k,
    covariance_type='full',
    random_state=42
)

clusters = best_gmm.fit_predict(X_pca)

# ======================
# 5. ĐÁNH GIÁ CUỐI
# ======================
final_silhouette = silhouette_score(X_pca, clusters)

print("\n=== FINAL EVALUATION ===")
print("Final Silhouette Score:", final_silhouette)
print("Final BIC:", best_gmm.bic(X_pca))

# ======================
# 6. GỘP KẾT QUẢ
# ======================
df_result = df.copy()
df_result['cluster'] = clusters

# ======================
# 7. PHÂN TÍCH ĐỊNH LƯỢNG
# ======================
print("\n=== Trung bình các đặc trưng theo cụm ===")
print(df_result.groupby('cluster').mean())

print("\n=== Thống kê CO(GT) theo cụm ===")
print(df_result.groupby('cluster')[target_col].describe())

print("\n=== Giá trị trung bình CO(GT) theo cụm ===")
print(df_result.groupby('cluster')[target_col].mean())

# ======================
# 8. TRỰC QUAN HÓA PCA
# ======================
plt.figure(figsize=(10, 6))

for i in range(optimal_k):
    plt.scatter(
        X_pca[clusters == i, 0],
        X_pca[clusters == i, 1],
        label=f'Cluster {i}',
        s=15,
        alpha=0.7
    )

plt.title(f"GMM Clustering (K={optimal_k}) using BIC")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()

# ======================
#import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(12,5))

# ===== BIC =====
ax[0].plot(list(k_range), bic_scores, marker='o')
ax[0].set_title("BIC vs Number of Clusters")
ax[0].set_xlabel("K")
ax[0].set_ylabel("BIC")
ax[0].grid(True)

# ===== Silhouette =====
ax[1].plot(list(k_range), sil_scores, marker='o')
ax[1].set_title("Silhouette Score vs Number of Clusters")
ax[1].set_xlabel("K")
ax[1].set_ylabel("Silhouette Score")
ax[1].grid(True)

plt.tight_layout()
plt.show()

# ======================
# 11. BOXPLOT CO(GT)
# ======================
plt.figure(figsize=(8,5))

plt.boxplot([
    df_result[df_result['cluster'] == i][target_col]
    for i in range(optimal_k)
])

plt.title("CO(GT) Distribution by Cluster")
plt.xlabel("Cluster")
plt.ylabel("CO(GT)")
plt.grid(True)

plt.show()
```

```python

# (a) REGRESSION ANALYSIS
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# DATA PREPARATION
X = df_scaled.drop(columns=['CO(GT)', 'Year'], errors='ignore')
y = df_scaled['CO(GT)']

pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X)

print("Original shape:", X.shape)
print("Reduced shape:", X_reduced.shape)

# HYPERPARAMETER TUNING (REGULARIZATION)
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance']
}

grid = GridSearchCV(
    KNeighborsRegressor(),
    param_grid,
    cv=5,
    scoring='r2'
)

grid.fit(X, y)

best_model = grid.best_estimator_

print("\n=== BEST PARAMETERS ===")
print(grid.best_params_)

# EXPERIMENTS WITH DIFFERENT SPLITS
ratios = [(0.2, "4:1"), (0.3, "7:3"), (0.4, "6:4")]

results = []

for test_size, ratio_name in ratios:

    # ---------------- ORIGINAL DATA ----------------
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )

    best_model.fit(X_train, y_train)

    train_pred = best_model.predict(X_train)
    test_pred = best_model.predict(X_test)

    results.append([
        ratio_name,
        "Original",
        np.sqrt(mean_squared_error(y_test, test_pred)),
        mean_absolute_error(y_test, test_pred),
        r2_score(y_test, test_pred),
        r2_score(y_train, train_pred)
    ])

    # ---------------- REDUCED DATA ----------------
    Xr_train, Xr_test, yr_train, yr_test = train_test_split(
        X_reduced, y, test_size=test_size, random_state=42
    )

    best_model.fit(Xr_train, yr_train)

    train_pred_r = best_model.predict(Xr_train)
    test_pred_r = best_model.predict(Xr_test)

    results.append([
        ratio_name,
        "Reduced",
        np.sqrt(mean_squared_error(yr_test, test_pred_r)),
        mean_absolute_error(yr_test, test_pred_r),
        r2_score(yr_test, test_pred_r),
        r2_score(yr_train, train_pred_r)
    ])

# RESULTS TABLE
result_df = pd.DataFrame(results, columns=[
    "Train/Test",
    "Data Type",
    "RMSE",
    "MAE",
    "R2_Test",
    "R2_Train"
])

print("\n=== KNN REGRESSION RESULTS ===")
print(result_df)
```

```python
#IMPORT + DATA
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
#CHUẨN BỊ DỮ LIỆU
target_col = "CO(GT)"

df_model = df.copy()

y_raw = df_model[target_col]
X_raw = df_model.drop(columns=[target_col], errors="ignore")

print(X_raw.shape, y_raw.shape)
#CHUẨN HÓA + PCA (1/3 số chiều)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)

n_components = max(1, X_raw.shape[1] // 3)

pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X_scaled)

print("Explained variance:", pca.explained_variance_ratio_)
#TẠO CLASSIFICATION LABEL (KẾT HỢP BIN + CLUSTER IDEA)


n_bins = 4

df_model["class"] = pd.qcut(
    df_model[target_col],
    q=n_bins,
    labels=False,
    duplicates="drop"
)
#LẤY NGƯỠNG (BOUNDARIES)
bins, edges = pd.qcut(
    df_model[target_col],
    q=n_bins,
    retbins=True,
    duplicates="drop"
)

intervals = pd.DataFrame({
    "Class": range(len(edges)-1),
    "Min": edges[:-1],
    "Max": edges[1:]
})

print(intervals)
#DATASET FINAL
y = df_model["class"]

X_original = X_scaled
X_reduced = X_pca
#HÀM TRAIN + EVALUATION
def run_models(X, y, name):

    results = []
    splits = {"80:20":0.2, "70:30":0.3, "60:40":0.4}

    for split_name, test_size in splits.items():

        X_train, X_test, y_train, y_test = train_test_split(
            X, y,
            test_size=test_size,
            random_state=42,
            stratify=y
        )

        # ======================
        # NAIVE BAYES
        # ======================
        nb = GaussianNB()
        nb.fit(X_train, y_train)
        pred_nb = nb.predict(X_test)

        results.append([
            name, split_name, "Naive Bayes",
            accuracy_score(y_test, pred_nb)
        ])

        # ======================
        # KNN
        # ======================
        knn = KNeighborsClassifier(n_neighbors=5, weights="distance")
        knn.fit(X_train, y_train)
        pred_knn = knn.predict(X_test)

        results.append([
            name, split_name, "KNN",
            accuracy_score(y_test, pred_knn)
        ])

    return results
#CHẠY THỰC NGHIỆM
results = []

results += run_models(X_original, y, "Original")
results += run_models(X_reduced, y, "PCA")

df_results = pd.DataFrame(
    results,
    columns=["Dataset", "Split", "Model", "Accuracy"]
)

print(df_results)
```

```python
#SO SÁNH KẾT QUẢ
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

sns.barplot(
    data=df_results[df_results["Split"]=="80:20"],
    x="Model",
    y="Accuracy",
    hue="Dataset"
)

plt.title("Model Comparison (80:20 split)")
plt.ylim(0,1)
plt.show()
#CONFUSION MATRIX (BEST MODEL)
best_X = X_reduced
best_y = y

X_train, X_test, y_train, y_test = train_test_split(
    best_X, best_y,
    test_size=0.2,
    random_state=42,
    stratify=best_y
)

model = KNeighborsClassifier(n_neighbors=5, weights="distance")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix (Best Model)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

print(classification_report(y_test, y_pred))
```
